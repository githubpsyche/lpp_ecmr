Plot SPC curves factored by categorical features (e.g., emotion tags).


```python
from pathlib import Path

from jaxcmr.analyses.cat_spc import plot_cat_spc
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data
import matplotlib.pyplot as plt
```

```python
data_name = "TalmiEEG"
data_query =  "data['subject'] > 0"
category_field = "condition"
contrast_name = "condition"
category_values = [1, 2]
labels = ["Negative", "Neutral"]
color_cycle = ["red", "black"]
ylim = (0, 1)
output_path = "work/reference_analyses/reference_catspc.png"
```

```python
project_root = Path(find_project_root())
data_path = project_root / "data" / f"{data_name}.h5"
output_file = Path(output_path)
if not output_file.is_absolute():
    output_file = project_root / output_file
output_file.parent.mkdir(parents=True, exist_ok=True)
output_path = str(output_file)
data = load_data(data_path)
trial_mask = generate_trial_mask(data, data_query)
```

```python
axis = plot_cat_spc(
    datasets=[data],
    trial_masks=trial_mask,
    category_field=category_field,
    category_values=category_values,
    contrast_name=contrast_name,
    labels=labels,
    color_cycle=color_cycle,
)
axis.set_ylim(ylim)
plt.savefig(output_path, bbox_inches="tight", dpi=600)
```
