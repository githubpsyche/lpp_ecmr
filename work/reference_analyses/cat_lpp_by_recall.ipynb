Plot the LPP amplitudes for items of a given category, separated by whether they were later recalled or not.

```python
from pathlib import Path
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data
from lpp_ecmr.data_contract import MIXED_TRIAL_QUERY
import matplotlib.pyplot as plt
from jaxcmr.analyses.cat_lpp_by_recall import plot_cat_lpp_by_recall
```

```python
data_name = "TalmiEEG"
data_query = MIXED_TRIAL_QUERY
category_field = "condition"
contrast_name = "Negative Items"
lpp_field = "EarlyLPP"
category_value = 1
labels = ["Recalled", "Unrecalled"]
color_cycle = None
ylim = None
output_path = "work/reference_analyses/cat_lpp_by_recall_EarlyLPP_Negative.png"
exclude_ci = False
```

```python
project_root = Path(find_project_root())
data_path = project_root / "data" / f"{data_name}.h5"
output_file = Path(output_path)
if not output_file.is_absolute():
    output_file = project_root / output_file
output_file.parent.mkdir(parents=True, exist_ok=True)
output_path = str(output_file)
data = load_data(data_path)
trial_mask = generate_trial_mask(data, data_query)
```

```python
axis = plot_cat_lpp_by_recall(
    datasets=[data],
    trial_masks=trial_mask,
    category_field=category_field,
    category_value=category_value,
    lpp_field=lpp_field,
    exclude_ci=exclude_ci,
    color_cycle=color_cycle,
    contrast_name=contrast_name,
    labels=labels,
)
axis.set_ylim(ylim)
plt.tight_layout()
plt.savefig(output_path, bbox_inches="tight", dpi=600)
```
