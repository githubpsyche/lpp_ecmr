# Render Model Comparisons

Generate comparison notebooks for each stop condition, at both subject-wise and group levels.

```python
import os
import papermill as pm
from jaxcmr.helpers import find_project_root

project_root = find_project_root()
workspace = os.path.dirname(project_root)
template = os.path.join(workspace, "jaxcmr", "templates", "model_comparison.ipynb")
rendered_dir = os.path.join(project_root, "analyses", "rendered")
os.makedirs(rendered_dir, exist_ok=True)
```

```python
_sc_ft = "50_set_likelihood_fixed_term_best_of_3"
_ecmr_ft = "50_set_likelihood_best_of_3"
_stop_ft = "50_set_likelihood_predicted_term_best_of_3"
_group_ft = "group_level_best_of_3"

_base_query_params = [
    "encoding_drift_rate", "start_drift_rate", "recall_drift_rate",
    "shared_support", "item_support", "learning_rate",
    "primacy_scale", "primacy_decay", "choice_sensitivity",
    "emotion_scale", "lpp_main_scale", "lpp_inter_scale",
]

_stop_query_params = _base_query_params + [
    "stop_probability_scale", "stop_probability_growth",
]

comparisons = [
    {
        "run_tag": "Exclude_Termination",
        "model_class_names": ["CMR", "LPP_CMR", "eCMR", "LPP_eCMR"],
        "model_display_titles": ["CMR", "LPP-CMR", "eCMR", "LPP-eCMR"],
        "model_fit_tags": [_sc_ft, _sc_ft, _ecmr_ft, _ecmr_ft],
        "nesting_pairs": [["CMR", "LPP-CMR"], ["eCMR", "LPP-eCMR"]],
        "query_parameters": _base_query_params,
    },
    {
        "run_tag": "PositionalStop",
        "model_class_names": ["CMR_PositionalStop", "LPP_CMR_PositionalStop", "eCMR_PositionalStop", "LPP_eCMR_PositionalStop"],
        "model_display_titles": ["CMR", "LPP-CMR", "eCMR", "LPP-eCMR"],
        "model_fit_tags": [_stop_ft] * 4,
        "nesting_pairs": [["CMR", "LPP-CMR"], ["eCMR", "LPP-eCMR"]],
        "query_parameters": _stop_query_params,
    },
    {
        "run_tag": "RatioStop",
        "model_class_names": ["CMR_RatioStop", "LPP_CMR_RatioStop", "eCMR_RatioStop", "LPP_eCMR_RatioStop"],
        "model_display_titles": ["CMR", "LPP-CMR", "eCMR", "LPP-eCMR"],
        "model_fit_tags": [_stop_ft] * 4,
        "nesting_pairs": [["CMR", "LPP-CMR"], ["eCMR", "LPP-eCMR"]],
        "query_parameters": _stop_query_params,
    },
    {
        "run_tag": "Group_Exclude_Termination",
        "model_class_names": ["CMR", "LPP_CMR", "eCMR", "LPP_eCMR"],
        "model_display_titles": ["CMR", "LPP-CMR", "eCMR", "LPP-eCMR"],
        "model_fit_tags": [_group_ft] * 4,
        "nesting_pairs": [["CMR", "LPP-CMR"], ["eCMR", "LPP-eCMR"]],
        "query_parameters": _base_query_params,
    },
    {
        "run_tag": "Group_PositionalStop",
        "model_class_names": ["CMR_PositionalStop", "LPP_CMR_PositionalStop", "eCMR_PositionalStop", "LPP_eCMR_PositionalStop"],
        "model_display_titles": ["CMR", "LPP-CMR", "eCMR", "LPP-eCMR"],
        "model_fit_tags": [_group_ft] * 4,
        "nesting_pairs": [["CMR", "LPP-CMR"], ["eCMR", "LPP-eCMR"]],
        "query_parameters": _stop_query_params,
    },
    {
        "run_tag": "Group_RatioStop",
        "model_class_names": ["CMR_RatioStop", "LPP_CMR_RatioStop", "eCMR_RatioStop", "LPP_eCMR_RatioStop"],
        "model_display_titles": ["CMR", "LPP-CMR", "eCMR", "LPP-eCMR"],
        "model_fit_tags": [_group_ft] * 4,
        "nesting_pairs": [["CMR", "LPP-CMR"], ["eCMR", "LPP-eCMR"]],
        "query_parameters": _stop_query_params,
    },
]
```

```python
for comp in comparisons:
    params = {
        "data_names": ["TalmiEEG"],
        "fit_dir": "fits/",
        "table_dir": "tables",
        "use_aicc": False,
        "data_path": "",
        "trial_query": "",
        "dataset_groups": {},
        **comp,
    }

    output = os.path.join(rendered_dir, f"model_comparison_{comp['run_tag']}.ipynb")
    print(output)

    pm.execute_notebook(
        template, output,
        parameters=params,
        cwd=project_root,
        progress_bar=True,
    )
```
